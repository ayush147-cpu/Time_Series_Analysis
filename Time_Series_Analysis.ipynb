# **Defines the size of Dataset according to our need**

```python
start_date="01-01-2002"
end_date="01-01-2010"
timelimt=(data.index >= start_date) & (data.index < end_date)
data1=data.loc[timelimt]
```

```python
data1.head()
```

```python
data1.tail()
```

```python
data1.plot(figsize=(12,8))
```

# **Visualize the Data on quarterly basis with the help of rolling function**

```python
data1quarterly=data.rolling(window=3).mean()
```

```python
data1quarterly.plot(figsize=(12,8))
```

# **Visualize the Data on half yearly basis with the help of rolling function**

```python
data1halfyearly=data.rolling(window=6).mean()
```

```python
data1halfyearly.plot(figsize=(12,8))
```

# **Visualize the Data on yearly basis with the help of rolling function**

```python
data1yearly=data.rolling(window=12).mean()
```

```python
data1yearly.plot(figsize=(12,8))
```
